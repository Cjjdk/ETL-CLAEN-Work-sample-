```python
# Keep the first occurrence and drop the rest
df = df.drop_duplicates(subset=["VIN"], keep='first')
```

```python

# or
# Standardize all column names: strip spaces and make uppercase
df.columns = df.columns.str.strip().str.upper()
```

```python
import pandas as pd

# List of numeric columns
numeric_cols = [
    'TRANSACTION_ID',
    'DEALER_REGION',
    'REPAIR_AGE',
    'KM',
    'COMPLAINT_CD_CSI',
    'ORD_SELLING_SRC_CD',
    'GLOBAL_LABOR_CODE',
    'CAMPAIGN_NBR',
    'REPORTING_COST',
    'TOTALCOST',
    'LBRCOST',
    'TRANSMISSION_SOURCE_PLANT',
    'SRC_TXN_ID',
    'SRC_VER_NBR',
    'TRANSACTION_CNTR',
    'LAST_KNOWN_DELVRY_TYPE_CD',
    'NON_CAUSAL_PART_QTY',
    'SALES_REGION_CODE'
]

# Detect and handle outliers using IQR method
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Option 1: Clip extreme values to bounds
    df[col] = df[col].clip(lower_bound, upper_bound)
    
    # Option 2 (Alternative): Remove rows with outliers
    # df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("Outliers handled for numeric columns.")
```
